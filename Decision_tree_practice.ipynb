```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
df =pd.read_csv("C:\\Users\\vamsi\\Downloads\\CarPrice_Assignment.csv")
df.head(5)
```

# Extract all Numbers from a String Column in Python Pandas

```python
df['car_model'] = df['CarName'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
```

```python
df.head()
```

```python
df.shape
```

```python
del df['car_ID']
```

```python
df[df.isnull().any(axis=1)].head()
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
df.head(3)
```

```python
clean_data=df.copy()
clean_data['price_update']=(clean_data['price']>16500)*1
clean_data['price_update'].head()
```

```python
y=clean_data[['price_update']].copy()
y.head()
```

```python
df.columns
```

```python
car_purchase={'symboling', 'wheelbase', 'carlength',
        'boreratio', 'stroke', 'compressionratio',
       'horsepower', 'peakrpm', 'citympg', 'highwaympg','car_model'}
```

```python
x=clean_data[car_purchase].copy()
x.columns
```

```python
y.columns
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=205)
```

```python
car_classifier=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
car_classifier.fit(x_train,y_train)
```

```python
y_predicted=car_classifier.predict(x_test)
```

```python
accuracy_score(y_test,y_predicted)*100
```

```python
confusion_matrix(y_test,y_predicted)
```
